```python
import arxiv
import pandas as pd
import datetime
import numpy as np
import json
from collections import OrderedDict
```

```python
#Fetch Research Paper Metadata using arXiv API
paper_data ={}

paper_categories = ['cs.AI','cs.CE','cs.SE','cs.DB','cs.NI','cs.GR']
no_of_papers_per_category = 5000
paper_id = 0
```

```python
def extract_fields(item):
    item_details = {}
    item_details['arxiv_id'] = item['id']
    item_details['published'] = item['published']
    item_details['title'] = item['title']
    item_details['summary'] = item['summary']
    item_details['authors'] = item['authors']
    item_details['arxiv_primary_category'] = item['arxiv_primary_category']['term']
    item_details['pdf_url'] = item['pdf_url']
    item_details['affiliation'] = item['affiliation']
    item_details['journal_reference'] = item['journal_reference']
    item_details['doi'] = item['doi']
    return item_details
```

```python
for i in range(len(paper_categories)):
    category_to_pull = paper_categories[i]
    print("category_to_pull : ", category_to_pull)
    string_to_search = "cat:" + str(category_to_pull)
    print("string_to_search : ", string_to_search)
    paper_response = arxiv.query(query=string_to_search,max_results=no_of_papers_per_category)
    print("number of papers fetched", len(paper_response))
    for i in range(len(paper_response)):
        print("Processing paper : ", i+1)
        detailed_fields = extract_fields(paper_response[i])
        paper_id = paper_id + 1
        paper_data[paper_id] = detailed_fields
```

```python
#Get paper metadata in list
paper_rows = []
for i in paper_data:
    paper_rows.append(paper_data[i])
```

```python
#Mappings data

import random
mappings={}

for sid in range(1,len(student_data)+1):
    papers_referred = []
    if sid <= 20 :
        for j in range(random.randint(0,20)):
            r = random.randint(1,4991)
            if r not in papers_referred: papers_referred.append(r)
    if sid >= 21 and sid <= 40 :
        for j in range(random.randint(0,20)):
            r = random.randint(5001,8146)
            if r not in papers_referred: papers_referred.append(r)
    if sid >= 41 and sid <=60 :
        for j in range(random.randint(0,20)):
            r = random.randint(8156,13146)
            if r not in papers_referred: papers_referred.append(r)
    if sid >= 61 and sid <= 80 :
        for j in range(random.randint(0,20)):
            r = random.randint(13155,17590)
            if r not in papers_referred: papers_referred.append(r)
    if sid >= 81 and sid <= 100 :
        for j in range(random.randint(0,20)):
            r = random.randint(17591,22590)
            if r not in papers_referred: papers_referred.append(r)
    if sid >= 101 and sid <= 120 :
        for j in range(random.randint(0,20)):
            r = random.randint(22591,24256)
            if r not in papers_referred: papers_referred.append(r)
    mappings[sid] = papers_referred

def add_item(r,i):
    existing = list(mappings[r])
    existing = existing + list([i])
    mappings[r] = list(existing)

ai_common_papers = [4991,4992,4993,4994,4995,4996,4997,4998,4999,5000]
for i in ai_common_papers:
    r = random.randint(1,20)
    add_item(r,i)
    if(i%2 == 0):
        add_item(r+20,i)
    add_item(r+100,i)
        
ce_common_papers = [8146,8147,8148,8149,8150,8151,8152,8153,8154,8155]
for i in ce_common_papers:
    r = random.randint(1,20)
    add_item(r,i)
    if(i%2 == 0):
        add_item(r+20,i)
    add_item(r+40,i)
    
se_common_papers = [13146,13147,13148,13149,13150,13151,13152,13153,13154,13155]
for i in se_common_papers:
    r = random.randint(1,20)
    if(i%2 == 0):
        add_item(r+20,i)
    add_item(r+40,i)
    if(i%2 == 0):
        add_item(r+60,i)
    add_item(r+80,i)
    

for stid in range(1,len(student_data)+1):    
    print("sid :", stid, "papers_referred: ", mappings[stid])
```
